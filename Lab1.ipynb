```python
import numpy as np
from google.colab.patches import cv2_imshow
import cv2

#2a. Read an image from the Drive
img = cv2.imread('/content/drive//My Drive/ColabNotebooks/victoria.jpeg')
#2b. Show the image
cv2_imshow(img)
```

```python
#2c. Print the shape of the image
print(img.shape)
```

```python
#2d. Convert the image into a greyscale image
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_grey)
```

```python
#2e. Crop the image as the half size 
img_half = img[0:img.shape[0]//2, 0:img.shape[1]//2, :]
cv2_imshow(img_half)
```

```python
#Save the image
cv2.imwrite('/content/drive//My Drive/ColabNotebooks/half.png', img_half)
```
