```python
import numpy as np
import cv2
from  matplotlib import pyplot as plt

#2a. Read an image from the Drive
img = cv2.imread('pictures/123.jpg')
#2b. Show the image
#cv2.imshow('image',img) #cv2.imgshow() 不能在jp book中展示图片
plt.imshow(img)
plt.show()
```

```python
#2c. Print the shape of the image
print(img.shape)
```

```python
#2d. Convert the image into a greyscale image
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_grey)
plt.show()
```

```python
#2e. Crop the image as the half size 
img_half = img[0:img.shape[0]//2, 0:img.shape[1]//2, :]
plt.imshow(img_half)
plt.show()
```

```python
#Save the image
cv2.imwrite('pictures/half.png', img_half)
```
